```python
# Import libraries

import pandas as pd
```

```python
# Iport summer data

oly_s = pd.read_csv("data/summer.csv")
oly_s.head ()
```

```python
oly_w = pd.read_csv("data/winter.csv")
oly_w.head ()
```

```python
oly_s.info()
```

```python
# Metjod 1 to return a Series. 
# df.column name

oly_w.Year
```

```python
# Metjod 2 to return a Series. NOTE: THIS IS THE PREFERRED METHOD. 
# df["Column name"] also returns the series

oly_w["Year"]
```

```python
# df multiple columns

df = oly_w[["Year", "City"]]
df.sample(10)
```

```python
df.info()
```

```python
# Sorting
# df_name.sort_values(by=["Value 1", "Value2"], ascending = True) For A-Z numbers to bigger numbers
# df_name.sort_values(by=["Value 1", "Value2"], ascending = False) For A-Z numbers or bigger to smaller numbers

ciy_yr_sort = df.sort_values(by=["City", "Year"], ascending = True)
ciy_yr_sort
```

```python
#The order matters. What you input fists is what will be prioritized 
ciy_yr_sort = df.sort_values(by=["Year", "City"], ascending = True)
ciy_yr_sort
```

```python
# Now in discending order
ciy_yr_sort = df.sort_values(by=["City", "Year"], ascending = False)
ciy_yr_sort
```

```python
# Questions to consider. Should winter ans summer remain separate? Depends on the question
# you're tryiing to answer

# Butn if we combine them, how can we tell them appart? We create a new column ang give it the same value
# for the entire data set

oly_s["Season"] = "Summer"
oly_s.head()
```

```python
oly_w["Season"] = "Winter"
oly_w.head()
```

```python
# Are the two column heather identicals?

list(oly_s.columns) == list (oly_w.columns)
```

```python
# join them

join_list = [oly_s, oly_w]
oly = pd.concat (join_list, sort = True)
oly
```

```python
# conirm the correct number of rows after the match

oly.shape [0] == oly_w.shape[0] + oly_s.shape[0]
```

```python
oly.sample(10)
```

```python
oly.to_csv("data/olympics.csv", encoding = "UTF-8", index = False)
```
